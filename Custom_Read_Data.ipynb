```python
import torch

from glob import glob
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd



class CustomReadData():
    
    #data_path = './rob535-fall2021-final-project-data/trainval/'
    #labels_path = './rob535-fall2021-final-project-data/trainval/labels.csv'
    
    def __init__(self, data_path, labels_path):
        self.data_path = data_path
        self.labels_path = labels_path
        
    def im_read(self):
        
        files = glob(self.data_path + '*/*_image.jpg')
        df = pd.read_csv(self.labels_path)
        
        gd_list = []
        
        for f in files:
            f_split = f.split("/")
            
            folder = f_split[2]    #check this line "\\" if error occurs and check the index values
            
            img_name = f_split[3].split("_")[0]
            
            file_name = folder + '/' + img_name

            gd = df.loc[df['guid/image'] == file_name]
            
            gd = gd.to_numpy().tolist()

            gd_list.append(gd[0][1])
            
        return files, gd_list

```
